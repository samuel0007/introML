```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([4, -0.9, 2.9,-3.4,4.9])
y = np.array([-6.9,2.3,8.7,7.5,-18.9])
```

```python
phi = np.array([[1 for el in x], [el for el in x], [el*el for el in x], [np.sin(el) for el in x]]).transpose()
w = np.linalg.lstsq(phi, y, rcond=None)[0]
f = lambda x: w[0] + w[1]*x + w[2]*x*x + w[3]*np.sin(x)
```

```python
x_ = np.linspace(-5, 5, 200)
plt.plot(x_, f(x_))
plt.plot(x, y, 'o')
print(w)
```
