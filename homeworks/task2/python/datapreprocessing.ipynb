### __Pre processing the CSV files so they can be directly loaded by the cpp files__



```python
import pandas as pd
```

```python
import pandas as pd
```

```python
train_x = pd.read_csv("../csv/train_features.csv")
train_y = pd.read_csv("../csv/train_labels.csv")
test_x = pd.read_csv("../csv/test_features.csv")
```

```python
train_x = train_x.groupby(train_x["pid"], sort=False, as_index=False).agg(
    "mean"
)
train_x.drop("Time", axis=1, inplace=True)

test_x = test_x.groupby(test_x["pid"], sort=False, as_index=False).agg("mean")
test_x.drop("Time", axis=1, inplace=True)
```

```python
train_x_mean = train_x.fillna(train_x.mean())
train_x_zero = train_x.fillna(0.0)

test_x_mean = test_x.fillna(test_x.mean())
test_x_zero = test_x.fillna(0)
```

```python
train_x_mean.to_csv("../csv/train_mean.csv", header=False, index=False)
train_x_zero.to_csv("../csv/train_zero.csv", header=False, index=False)
train_y.to_csv("../csv/train_labels_noHeader.csv", header=False, index=False)
test_x_mean.to_csv("../csv/test_mean.csv", header=False, index=False)
test_x_zero.to_csv("../csv/test_zero.csv", header=False, index=False)
```

```python
classification_labels = [
    "LABEL_BaseExcess",
    "LABEL_Fibrinogen",
    "LABEL_AST",
    "LABEL_Alkalinephos",
    "LABEL_Bilirubin_total",
    "LABEL_Lactate",
    "LABEL_TroponinI",
    "LABEL_SaO2",
    "LABEL_Bilirubin_direct",
    "LABEL_EtCO2",
]
train_y = train_y[classification_labels]
train_y.to_csv("train_classification_labels.csv", header=False, index=False)
```
