### SVM in Python




```python
import pandas as pd
import numpy as np
```

```python
train_x = pd.read_csv("../csv/train_features.csv")
train_y = pd.read_csv("../csv/train_labels.csv")
test_x = pd.read_csv("../csv/test_features.csv")
```

```python
train_x = train_x.groupby(train_x["pid"], sort=False).agg("mean")
train_x.drop("Time", axis=1, inplace=True)
train_x.fillna(0, inplace=True)

test_x = test_x.groupby(test_x["pid"], sort=False).agg("mean")
test_x.drop("Time", axis=1, inplace=True)
test_x.fillna(0, inplace=True)
```

```python
train_x.head()
```

```python
classification_labels = [
    "LABEL_BaseExcess",
    "LABEL_Fibrinogen",
    "LABEL_AST",
    "LABEL_Alkalinephos",
    "LABEL_Bilirubin_total",
    "LABEL_Lactate",
    "LABEL_TroponinI",
    "LABEL_SaO2",
    "LABEL_Bilirubin_direct",
    "LABEL_EtCO2",
    "LABEL_Sepsis",
]
train_y = train_y[classification_labels]
```

```python
train_y.head()
```

```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn import svm

rbf_svc = svm.SVC(kernel="rbf")
classifier = MultiOutputClassifier(rbf_svc).fit(train_x, train_y)
classifier.predict_proba(test_x)
```
