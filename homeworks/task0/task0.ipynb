```python
import pandas as pd
from sklearn.metrics import mean_squared_error

df = pd.read_csv("train.csv", index_col="Id")
df.head()
```

```python
atol = 1e-12
y = df.loc[:, "y"]
y_pred = df.loc[:, df.columns != "y"].mean(axis=1)
assert ((y - y_pred) < atol).all()

print(mean_squared_error(y, y_pred) ** 0.5)
```

```python
df = pd.read_csv("test.csv", index_col="Id")
sol = pd.DataFrame(df.mean(axis=1), columns=["y"])
sol.to_csv("solution.csv")
```
