```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")
df_target = df.iloc[:, 0]
df_data = df.iloc[:, 1:]
df.head()
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

param_grid = {"alpha": [0.1, 1, 10, 100, 200]}
clf = GridSearchCV(
    Ridge(), param_grid, scoring="neg_root_mean_squared_error", cv=10
)
clf.fit(df_data, df_target)
pd.DataFrame(-clf.cv_results_["mean_test_score"]).to_csv(
    "solution.csv", index=False, header=False
)
```
