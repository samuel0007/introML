```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Subtask 1

```python
train_x = pd.read_csv("train_features.csv")
train_y = pd.read_csv("train_labels.csv")
test_x = pd.read_csv("test_features.csv")
```

```python
classification_labels = [
    "pid",
    "LABEL_BaseExcess",
    "LABEL_Fibrinogen",
    "LABEL_AST",
    "LABEL_Alkalinephos",
    "LABEL_Bilirubin_total",
    "LABEL_Lactate",
    "LABEL_TroponinI",
    "LABEL_SaO2",
    "LABEL_Bilirubin_direct",
    "LABEL_EtCO2",
]
train_y = train_y[classification_labels]
train_y.head()
```

```python
train_x = train_x.groupby("pid")
train_x_count = train_x.count()
# Data is super sparse
```

```python
train_x_count.head()
train_x_count.iloc[[0]]
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import RidgeClassifierCV
from sklearn.preprocessing import StandardScaler
from sklearn.multioutput import MultiOutputClassifier
from sklearn.pipeline import make_pipeline
from sklearn.metrics import roc_auc_score
from tqdm import tqdm

# Always scale the input. The most convenient way is to use a pipeline.
SGDClf = make_pipeline(
    StandardScaler(), SGDClassifier(max_iter=1000, tol=1e-3)
)
# RidgeClfCV = make_pipeline(StandardScaler(),
#                     RidgeClassifierCV())

models = [SGDClf]
# models = [RidgeClfCV]
classifiers = []
for model in tqdm(models):
    classifiers.append(
        MultiOutputClassifier(model).fit(train_x_count, train_y)
    )
```
